```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = 14, 12
rcParams['font.size'] = 16
rcParams['axes.labelsize'] = 14
rcParams['xtick.labelsize'] = 13
rcParams['ytick.labelsize'] = 13
rcParams['legend.fontsize'] = 15

import seaborn as sns
sns.set_style("whitegrid")
```

```python
df = pd.read_csv('test.csv')
```

```python
df = df[(df['Transaction Time'] >= '2020-01-15') & (df['Transaction Time'] <= '2020-02-15')]
```

```python
df = df.sort_values('Transaction Time')
df = df.reset_index(drop=True)
```

```python
df = pd.concat([df, pd.get_dummies(df['Item Type'])], axis=1)
df = pd.concat([df, pd.get_dummies(df['Marketplace Currency'])], axis=1)
```

```python
currs = list(Counter(df['Marketplace Currency']))
```

```python
token = "af3e573c70cf4da2a27ab1e42e1c7557"
```

```python
from openexchangerate import OpenExchangeRates
client = OpenExchangeRates(token)
# print(client.currencies())
```

```python
import re
pattern = ' \S*'
df['general_timestamp'] = df['Transaction Time'].apply(lambda x: re.sub(pattern, '', x))
timestamp = df['general_timestamp'].unique()
timestamp.sort()
```

```python
dictionary = {}
def get_historical_data():
    for time in timestamp:
        inter_currency = client.historical(f"{time}")[0]
        for curr in currs:
            dictionary[f'{curr}_{time}'] = inter_currency[curr]
```

```python
get_historical_data()
```

```python
import numpy as np
def sum_in_usd(timestamp, money, count, currency):
    return money * count / dictionary[f'{currency}_{timestamp[:10]}']

df['money in usd'] = np.vectorize(sum_in_usd)(df['Transaction Time'], df['Sales Price (Marketplace Currency)'],
                         df['Units'], df['Marketplace Currency'])
```

```python
df_agg = df.groupby(['Title']).sum()
```

```python
df_agg['avg price per game'] = df_agg['money in usd'] / df_agg['Units']
```

```python
country_list = ['Canada', 'Europe', 'England_other', 'Japan', 'US_other']
def get_best_country(*args):
    return country_list[np.argmax(args)]
df_agg['Country'] = np.vectorize(get_best_country)(df_agg['CAD'], df_agg['EUR'], df_agg['GBP'], df_agg['JPY'], df_agg['USD'])
```

# Нарисуем пару графиков для понимания картины в целом, хотя данных за месяц
# можно попасть в сезонность(

```python
def get_historical_price(game):
    return df.groupby(['Title', 'general_timestamp']).sum().reset_index(level=1).loc[game][['general_timestamp', 'money in usd']]
```

```python
dynamic_price = []
dynamic_title = df_agg.sort_values('money in usd', ascending=False).head(5).index
for game in dynamic_title:
    dynamic_price.append(get_historical_price(game).values)
```

```python
timestamp = list(timestamp)
X_strip = []
for prices in dynamic_price:
    X_strip.append(timestamp.index(prices[:, 0].min()))
```

```python
def get_data(i): 
    return df[df['Title'] == dynamic_title[i]].groupby('Marketplace Currency').size().reset_index('Marketplace Currency').sort_values(0, ascending=False).head(3)
```

## Посмотрим топ-5 игр по прибыли, кто же в них играет, хотя скорее где.

```python
fig, axes = plt.subplots(6, 1, figsize=(12*5, 7*5), subplot_kw=dict(aspect="equal"), dpi= 80)

def func(pct, allvals):
    
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}% ({:d})".format(pct, absolute)



for i in range(6):
    if i !=5:
        data = get_data(i)[0]
        categories = get_data(i)['Marketplace Currency']
    if i == 5:
        dat = df.groupby('Marketplace Currency').size().reset_index('Marketplace Currency')
        dat = dat.sort_values(0, ascending=False).head(3)
        data = dat[0]
        categories = dat['Marketplace Currency']
#         print(categories)
    # print(categories)
    explode = [0.1, 0.15, 0.2]

    wedges, texts, autotexts = axes[i].pie(data, 
                                      autopct=lambda pct: func(pct, data),
                                      textprops=dict(color="black"), 
                                      colors=plt.cm.Dark2.colors,
                                     startangle=100,shadow=True, 

                                     explode=explode)

    # Decoration
    axes[i].legend(wedges, categories, title="Currency", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.setp(autotexts, size=15, weight=700)
    if i != 5:
        axes[i].set_title(f"{dynamic_title[i]}")
    else:
        axes[i].set_title('Union destribution')
plt.show()
```
